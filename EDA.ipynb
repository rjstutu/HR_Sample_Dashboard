```python
import os
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cwd = os.getcwd()
data_file = f"{cwd}/input_data/raw_hr_data.csv"
df_hr = pd.read_csv(data_file)
df_hr.head()
```

## Get a glimpse of data, understand the data size and format.

```python
# df_hr.info()
df_hr.describe()
```

```python


plt.figure(figsize=(8, 6))
sns.countplot(x='Department', data=df_hr)
plt.title('Department Distribution')
plt.xlabel('Department')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=df_hr)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```
